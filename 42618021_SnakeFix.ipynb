```python
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=30,
      width_shift_range=0.4, # Shifting image width by 40%
      height_shift_range=0.4,# Shifting image height by 40%
      shear_range=0.2,       # Rotation across X-axis by 20%
      zoom_range=0.3,        # Image zooming by 30%
      vertical_flip = True,
      horizontal_flip=True)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (150, 200),
    class_mode = 'categorical',
    batch_size = 32
)

test_datagen = ImageDataGenerator(
      rescale=1./255)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size = (150, 200),
    class_mode = 'categorical',
    batch_size = 32
)
```

```python
from keras.layers import Dense, Dropout, Flatten
model = tf.keras.models.Sequential([
    
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 200, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    Dropout(0.25),

    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    Dropout(0.25),
    
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    Dropout(0.25),
    
    tf.keras.layers.Flatten(),
    
    tf.keras.layers.Dense(64, activation='relu'),
    Dropout(0.25),
    tf.keras.layers.Dense(17, activation='softmax')
])
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy',
              optimizer = 'adam',
              metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=30)
```

```python
history = model.fit_generator(train_generator,
                              epochs=200,
                              verbose=1,
                              validation_data=test_generator,
                              callbacks = [es])
```

```python
print("Accuracy of the model on train data is {:.2f}%".format(history.history["accuracy"][-1]*100))
```

```python
import matplotlib.pyplot as plt
acc=history.history['accuracy']
val_acc=history.history['val_accuracy']
loss=history.history['loss']
val_loss=history.history['val_loss']

epochs=range(len(acc))

fig = plt.figure(figsize=(14,7))
plt.plot(epochs, acc, 'r', label="Training Accuracy")
plt.plot(epochs, val_acc, 'b', label="Validation Accuracy")
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc='lower right')
plt.show()
```

```python
fig2 = plt.figure(figsize=(14,7))
plt.plot(epochs, loss, 'r', label="Training Loss")
plt.plot(epochs, val_loss, 'b', label="Validation Loss")
plt.legend(loc='upper right')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and validation loss')
```
